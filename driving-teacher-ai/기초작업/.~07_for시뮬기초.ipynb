```python
import pandas as pd
```

```python
data = pd.read_excel('./raw_data/요일별_연수_데이터3.xlsx')
data
```

```python
data['예약날짜'] = pd.to_datetime(data['예약날짜'])
data['학원도착날짜'] = pd.to_datetime(data['학원도착날짜'])
data['term'] = data['학원도착날짜'] - data['예약날짜']
data
```

```python
data_term = data.sort_values('term').reset_index(drop=True)
```

```python
data_term['term']
```

```python
import matplotlib.pyplot as plt
data_term['term'].plot()
```

```python
samples = []
for i, tick in enumerate(data_term['term']):
    print(i, tick, tick.total_seconds())
    if i >= 51:
        samples.append(tick.total_seconds())
```
