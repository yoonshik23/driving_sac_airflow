```python
import pandas as pd

# 주어진 시간 문자열
time_str = '2011-05-22 12:00:00'

# 문자열을 datetime 형식으로 변환하고 UTC로 설정
utc_time = pd.to_datetime(time_str).tz_localize('UTC')

# UTC 시간을 KST로 변환
kst_time = utc_time.tz_convert('Asia/Seoul')

# 결과 출력
print(kst_time)
```
